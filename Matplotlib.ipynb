```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a simple DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Scores": [85, 90, 78]
}
df = pd.DataFrame(data)

# Plot using Pandas
df.plot(x="Name", y="Scores", kind="bar")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Simple plot
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title("Example Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 35, 20, 20]  # Percentages or quantities
colors = ['red', 'yellow', 'pink', 'brown']  # Custom colors
explode = (0.1, 0, 0, 0)  # Explode the first slice (Apples)

# Create the pie chart
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)

# Add a title
plt.title("Fruit Distribution")

# Show the plot
plt.show()
```
